**3. Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## B. Penanganan terhadap Data

**1. nilai yang hilang (Missing Values)**

*   Rating (Imputasi)

```python
df['Rating'].dropna().describe()
```

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].dropna().mean())
```

*   Price (Imputasi)

```python
df['Price'].dropna().describe()
```

```python
df['Price'] = df['Price'].fillna(df['Price'].dropna().mean())
```

**2. Duplicated Values (Dihapus semua duplikasi)**

```python
df = df.drop_duplicates()
```

```python
df
```

**3. Outliers (Melakukan imputasi)**

```python
columns_to_impute = ["Rating", "Price"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

Melakukan Pengecekan Ulang

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

##  konstruksi data (Construct Data) (penambahan kolom)

**Penambahan Kolom: Kategori harga**

```python
df_cleaned = df.copy()
```

```python
df_cleaned["Price"] = pd.to_numeric(df_cleaned["Price"], errors="coerce")
```

```python
def kategori_harga(Price):
    if Price < 150:
        return "Murah"
    elif 150 <= Price <= 400:
        return "Menengah"
    else:
        return "Mahal"
```

```python
df_cleaned["kategori_Harga"] = df_cleaned["Price"].apply(kategori_harga)
```

```python
print(df_cleaned[["Price", "kategori_Harga"]])
```

**Penambahan Kolom: Kategori harga (dari rating/penilaian 1 sampai 5 menjadi 1 sampai 100)**

```python
if "Rating" in df_cleaned.columns:
    df_cleaned["Rating_yang_disesuaikan"] = df_cleaned["Rating"] * 20
```

```python
print(df_cleaned[["Rating", "Rating_yang_disesuaikan"]])
```
